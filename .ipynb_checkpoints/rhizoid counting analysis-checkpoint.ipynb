```python
rhizoid_data = pd.read_excel("rhizoid counting combined.xlsx")
# del rhizoid_data['sample']
# rhizoid_data.sort("number of rhizoids")
# rhizoid_data.sort("line")
```

```python
x = sns.factorplot("line","number of rhizoids",hue= "Treatment", data = rhizoid_data, kind = 'bar', aspect = 2, hue_order = ("OE","KD","Wt"))
plt.xticks(rotation=50)
```

```python
only_OE = rhizoid_data[rhizoid_data["Treatment"]!="KD"]
CaMV_bar_colors = [ "#4C72B0", "#4C72B0", "#4C72B0", "#4C72B0", "#4C72B0", "#C44E52"]
x = sns.barplot("line","number of rhizoids", data = only_OE, color=sns.color_palette(CaMV_bar_colors))
plt.suptitle('OE rhizoids after one day')
plt.xticks(rotation=-50)
#plt.savefig("C:/Users/Admin/Dropbox/Postgrad/35SMpSG9 rhizoids.png", dpi=1200)
```

```python
only_KD = rhizoid_data[rhizoid_data["line"]!="SpKD1 pHART A"]
only_KD = only_KD[only_KD["line"]!="SpKD1 pHART G"]
only_KD = only_KD[only_KD["Treatment"]!="OE"]
CaMV_KD_bar_colors = [ "#64B5CD", "#64B5CD", "#64B5CD", "#64B5CD","#64B5CD", "#C44E52"]
x = sns.barplot("line","number of rhizoids", data = only_KD, color=sns.color_palette(CaMV_KD_bar_colors))
plt.suptitle('35SKD rhizoids after one day')
plt.xticks(rotation=-50)
#plt.savefig("C:/Users/Admin/Dropbox/Postgrad/35SKD rhizoids.png", dpi=1200)
```
