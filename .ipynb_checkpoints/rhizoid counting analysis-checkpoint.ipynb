```python
rhizoid_data = pd.read_excel("rhizoid counting combined.xlsx")
rhizoid_data
```

```python
x = sns.factorplot("line","number of rhizoids",hue= "Treatment", data = rhizoid_data, kind = 'bar', aspect = 2, hue_order = ("OE","KD","Wt"))
plt.xticks(rotation=50)
```

```python
#examine data for normality
def hist(line):
    c = rhizoid_data[rhizoid_data["line"] == line]
    c = c["number of rhizoids"]
    c=np.array(c)
    plt.hist(c,bins=4)
    plt.title(line)
    plt.figure()

g = set(rhizoid_data["line"])
for line in g:
    hist(line)
```

```python
# calculate ranksums statistics
def ranksums(line):
    a = rhizoid_data[rhizoid_data["line"] == "Wt"]
    a = a["number of rhizoids"]
    b = rhizoid_data[rhizoid_data["line"] == line]
    b = b["number of rhizoids"]
    return sp.stats.ranksums(a,b)

g = set(rhizoid_data["line"])
g.remove("Wt")

for line in g:
    print(line, ranksums(line))
```
