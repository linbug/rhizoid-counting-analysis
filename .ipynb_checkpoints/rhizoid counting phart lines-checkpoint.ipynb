```python
import pandas as pd
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
rhizoid_data = pd.read_excel("rhizoid counting pHART.xlsx")
```

```python
# Add a treatment column

def define_treatment(x):
    if x[0] == 'M':
        return 'OE'
    elif x[0] == 'S':
        return 'KD'
    else:
        return 'Wt'
    
rhizoid_data['Treatment']=rhizoid_data['Line'].apply(define_treatment) 
rhizoid_data
```

```python
x = sns.factorplot("Line","number of rhizoids",hue= "Treatment", data = rhizoid_data, kind = 'bar', aspect = 1, hue_order = ("OE","KD","Wt"))
plt.xticks(rotation=-50)
```

```python
only_OE = rhizoid_data[rhizoid_data["Treatment"]!="KD"]
CaMV_bar_colors = [ "#55A868", "#55A868", "#55A868","#C44E52"]
x = sns.barplot("Line","number of rhizoids",data = only_OE, color=sns.color_palette(CaMV_bar_colors))
plt.suptitle('MpEFOE rhizoids')
plt.xticks(rotation=-50)
#plt.savefig("C:/Users/Admin/Dropbox/Postgrad/MpEFOE rhizoids.png", dpi =1200)
```

```python
only_KD = rhizoid_data[rhizoid_data["Treatment"]!="OE"]
pHART_bar_colors = [ "#7CC193", "#7CC193","#7CC193","#C44E52"]
x = sns.barplot("Line","number of rhizoids",data = only_KD, color=sns.color_palette(pHART_bar_colors))
plt.suptitle('35SKD rhizoids')
plt.xticks(rotation=-50)
#plt.savefig("C:/Users/Admin/Dropbox/Postgrad/MpEFKD rhizoids.png", dpi =1200)
```
